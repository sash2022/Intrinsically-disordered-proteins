```python
#step1: read the data set to choose sequeces out of it
import pandas as pd
import csv
import subprocess

# Load the excel file
input_file_path = '41592_2023_2159_MOESM4_ESM.xlsx'  # Replace with your actual file path
output_file_path = 'features.xlsx'  # Path to save the final features

# Specify the column name you want to display
column_name = ' sequence'  # Replace with your actual column name

# Name or index of the sheet you want to read
sheet_name = 'Table_S8'  # Replace with the actual sheet name or use an index (e.g., 0 for the first sheet)

# Read the specific sheet
data = pd.read_excel(input_file_path, sheet_name=sheet_name)
```

```python
#step2: run the Keresten code on the sequences from dataset to get their features 
import numpy as np

# Remove the identifier (e.g., 'Q99457_1_97') and flatten the list if necessary
def extract_values(line):
    return [float(x) for x in line if isinstance(x, str) and x.replace('.', '', 1).replace('-', '', 1).isdigit()]

feature_values=np.zeros((137,11))

# Loop through each sequence in the column ( read sequences and run the code and calculate the features. thankGOD!

for index, sequence in data[column_name].items():
   
    # Ensure the sequence is a string
    sequence = str(sequence)
    #print(f"Processing sequence at index {index}: {sequence}")

    # Run the external Python script with the sequence
    result = subprocess.run(['python', 'IDRlab_modified3.py', sequence],capture_output=True, text=True)  # Adjust 'IDRlab_modified.py' to your script name


    output_lines = [line for line in result.stdout.splitlines() if line.strip()]
    


    feature_values_4 = output_lines[3].split()  # Extracts the values in the second row. thankGOD!
    feature_values_6= output_lines[5].split()
    
    feature_values_4 = extract_values(feature_values_4)
    feature_values_6 = extract_values(feature_values_6)
    
    feature_values[index]=  feature_values_4 + feature_values_6
print(feature_values)
```

```python
# Step 3: Combine both header and resulted feature lines into a DataFrame
headers = ['fK', 'fR', 'fE', 'fD', 'fARO', 'mean_lambda', 'SHD', 'SCD', 'kappa', 'FCR', 'NCPR']  # Adjust header names as needed
print(len(headers))
# Initialize an empty DataFrame for storing features
output_features=pd.DataFrame()

for row in feature_values:
    df_row = pd.DataFrame([row], columns=headers) # Convert each row into a DataFrame
    output_features = pd.concat ([output_features, df_row], ignore_index=True)  # Append to main DataFrame

print (output_features)
```

```python
# Step 4: Write the df to Excel
output_file_path = "features.xlsx"
output_features.to_excel(output_file_path, index=False)

print(f"Data has been written to {output_file_path}")
```
